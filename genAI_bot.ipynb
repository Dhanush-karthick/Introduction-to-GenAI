```python
driver.get("https://morganstanley.tal.net/vx/candidate/apply/17979")
print(driver.page_source)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
titles = [event.text for event in soup.find_all('div',class_='form-control-static')]
print(titles)
```

```python
from langchain_groq import ChatGroq
titles_string = "\n".join(titles)
llm = ChatGroq(
    model="llama3-8b-8192",
    temperature=0,
    groq_api_key="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"
)

def handle_query(user_query):
    messages = [
        ("system",'''if query not related to morgan stanley just return:
        {query not appropriate}'''),
        ("system", '''answer to user according to morgan stanley summer annalyst
         intership '''),
        ("user", user_query),
        ("user", titles_string)
    ]


    ai_msg = llm.invoke(messages)
    return ai_msg.content

user_query = "When is this intership and who can apply"
response = handle_query(user_query)
print(response)
```
